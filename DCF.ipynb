# Discounted Cash Flow Model

## FCFF Method

```python
import yfinance as yf
import pandas as pd
import numpy as np

folder_path=r"C:\Users\nisha\Desktop\2026 PS-II\Offshoot courses\ECON F355 - BAV\Assignment\pythonProject\data"
FCFF=pd.read_csv(folder_path+'\\'+'FCFF.csv',index_col=0)

FCFF=FCFF.T
CF=pd.DataFrame()
CF["Cash Flows"]=FCFF['FCFF']

print(CF.T)
```

```python
costofcapital=pd.read_csv(folder_path+'\\'+'WACC.csv',index_col=0)

terminal_growth_rate=costofcapital.iloc[-1,0]
WACC=costofcapital.iloc[-5,0]


years = np.arange(1, len(CF) + 1) 
discount_factors = np.power(1 + WACC, years)
CF['Present Value'] = CF.iloc[:, 0] / discount_factors
    
print(CF)
```

```python
terminal_value=CF.iloc[-1,0]*(1+terminal_growth_rate)/(WACC-terminal_growth_rate)
print(terminal_value)

Discounted_terminal_value=terminal_value/discount_factors[-1]
print(Discounted_terminal_value)
```

```python
EV=Discounted_terminal_value+CF["Present Value"].sum()
print(EV)
```

```python
results_FCFF=pd.DataFrame()
results_FCFF.index=['Value']
results_FCFF["terminal_value"]=terminal_value
results_FCFF["Discounted_terminal_value"]=Discounted_terminal_value
results_FCFF["Enterprise Value"]=EV

bal_sheet=pd.read_csv(folder_path+'\\'+'Balance_sheet_req_data.csv',index_col=0)
mvd=bal_sheet.iloc[0,0]
cash=bal_sheet.iloc[-2,0]


mve=EV-mvd+cash
results_FCFF["MVD"]=mvd
results_FCFF["Slack Cash"]=cash
results_FCFF["MVE"]=mve

print(results_FCFF)
CF.to_csv(folder_path+'\\'+'FCFF_CFs.csv')
results_FCFF.to_csv(folder_path+'\\'+'FCFF_results.csv')
```
